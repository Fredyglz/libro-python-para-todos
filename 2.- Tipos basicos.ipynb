```python
# Jose Alfredo Romero Gonzalez
# 30/12/2019 01:51
# fredyglz.fr@gmail.com
# python 3.7.3

#TIPOS BASICOS

# esto es una cadena
c = "Hola mundo"

# y esto es un entero
e = 23

# podemos comprobarlo con la funcion type
print(type(c))
print(type(e))

'''
NUMEROS
'''
#Enteros
# print(type(entero)) devolvera int
entero = 23
print(type(entero))
#Ya no funciona 23L para declararlo como long
#entero = 23L 
#print(type(entero))
# 027 octa = 23 en base 10
#entero = 027
#print(type(entero))
# 0x17 hezadecimal = 23 en base 10
#entero = 0x17

#Reales
real = 0.2703
print(type(real))
real = 0.1e-3
print(real)
print(type(real))

#Complejos
complejo = 2.1+7.8j
print(complejo)
print(type(complejo))

#Operadores
r = 3 + 2   # r es 5 (suma)
print(r)
r = 4 - 7   # r es -3 (resta)
print(r)
r = -7      # r es -7 (negacion)
print(r)
r = 2 * 6   # r es 12 (multiplicacion)
print(r)
r = 2 ** 6  # r es 64 (exponente)
print(r)
r = 3.5 / 2 # r es 1.75 (division)
print(r)
r = 3.5//2  # r es 1.0 (division entera)
print(r)
r = 7 % 2   # r es 1 (modulo)
print(r)

#Operadores a nivel de bit
print("Operadores a nivel de bit")
r = 3 & 2 # r es 2
print(r)
r = 3 | 2 # r es 3
print(r)
r = 3 ^ 2 # r es 1
print(r)
r = ~3    # r es -4 0011
print(r)
r = 3 << 1# r es 6 0011 << 0110
print(r)
r = 3 >> 1# r es 1 0011 >> 0001
print(r)

'''
CADENAS
'''
print("CADENAS")
unicode = u"aóe"
raw  = r"\n"
print(unicode)
print(raw)

triple = '''primer linea
            esto se vera en otra linea'''
print(triple)

a = "uno "
b = "dos"
c = a + b # c es "unodos"
print(c)
c = a * 3 # c es "unounouno"
print(c)

'''
BOOLEANOS
'''
print("BOOLEANOS")
r = True and False  # r es False
print(r)
r = True or False   # r es True
print(r)
r = not True        # r es False
print(r)
r = 5 == 3          # r es False
print(r)
r = 5 != 3          # r es True
print(r)
r = 5 < 3           # r es False
print(r)
r = 5 > 3           # r es True
print(r)
```
